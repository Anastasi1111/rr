```python
import os
import shutil
import subprocess

def create_file(filename):
    with open(filename, 'w') as file:
        file.write('')  # Создание пустого файла
    print(f"Файл '{filename}' создан")

def delete_file(filename):
    if os.path.exists(filename):
        os.remove(filename)
        print(f"Файл '{filename}' удален")
    else:
        print(f"Файл '{filename}' не существует")

def move_file(source, destination):
    if os.path.exists(source):
        shutil.move(source, destination)
        print(f"Файл '{source}' перемещен в '{destination}'")
    else:
        print(f"Файл '{source}' не существует")

def copy_file(source, destination):
    if os.path.exists(source):
        shutil.copy(source, destination)
        print(f"Файл '{source}' скопирован в '{destination}'")
    else:
        print(f"Файл '{source}' не существует")

def create_folder(foldername):
    os.makedirs(foldername, exist_ok=True)
    print(f"Папка '{foldername}' создана")

def delete_folder(foldername):
    if os.path.exists(foldername):
        shutil.rmtree(foldername)
        print(f"Папка '{foldername}' удалена")
    else:
        print(f"Папка '{foldername}' не существует")

def move_folder(source, destination):
    if os.path.exists(source):
        shutil.move(source, destination)
        print(f"Папка '{source}' перемещена в '{destination}'")
    else:
        print(f"Папка '{source}' не существует")


def copy_folder(source, destination):
    if os.path.exists(source):
        shutil.copytree(source, destination)
        print(f"Папка '{source}' скопирована в '{destination}'")
    else:
        print(f"Папка '{source}' не существует")

def initialize_git_repo(folder):
    subprocess.run(['git', 'init', folder])

def commit_changes(folder, message):
    subprocess.run(['git', 'add', '-A'], cwd=folder)
    subprocess.run(['git', 'commit', '-m', message], cwd=folder)

def main():
    while True:
        print("\n==== Меню проводника ====")
        print("1. Создать файл")
        print("2. Удалить файл")
        print("3. Переместить файл")
        print("4. Копировать файл")
        print("5. Создать папку")
        print("6. Удалить папку")
        print("7. Переместить папку")
        print("8. Копировать папку")
        print("9. Инициализировать Git-репозиторий")
        print("10. Закоммитить изменения в Git-репозитории")
        print("11. Зарегистрироваться на GitHub и выгрузить программу")
        print("0. Закрыть программу")
        choice = input("Ваш выбор: ")
 
        if choice == "1":
            filename = input("Введите имя файла: ")
            create_file(filename)
     
        elif choice == "2":
            filename = input("Введите имя файла: ")
            delete_file(filename)
       
        elif choice == "3":
            source = input("Введите имя файла/папки для перемещения: ")
            destination = input("Введите путь для перемещения файла/папки: ")
            move_file(source, destination)
       
        elif choice == "4":
            source = input("Введите имя файла/папки для копирования: ")
            destination = input("Введите путь для сохранения копии файла/папки: ")
            copy_file(source, destination)
        
        elif choice == "5":
            foldername = input("Введите имя папки: ")
            create_folder(foldername)

        elif choice == "6":
            foldername = input("Введите имя папки: ")
            delete_folder(foldername)
       
        elif choice == "7":
            source = input("Введите имя папки для перемещения: ")
            destination = input("Введите путь для перемещения папки: ")
            move_folder(source, destination)
       
        elif choice == "8":
            source = input("Введите имя папки для копирования: ")
            destination = input("Введите путь для сохранения копии папки: ")
            copy_folder(source, destination)
        
        elif choice == "9":
            folder = input("Введите путь к папке, в которой будет создан Git-репозиторий: ")
            initialize_git_repo(folder)
       
        elif choice == "10":
            folder = input("Введите путь к Git-репозиторию: ")
            message = input("Введите сообщение коммита: ")
            commit_changes(folder, message)
 
        elif choice == "11":
            username = input("Введите ваше имя пользователя на GitHub: ")
            email = input("Введите вашу электронную почту, связанную с аккаунтом на GitHub: ")
            subprocess.run(['git', 'config', '--global', 'user.name', username])
            subprocess.run(['git', 'config', '--global', 'user.email', email])
            repo_name = input("Введите имя репозитория на GitHub, куда вы хотите загрузить программу: ")
            subprocess.run(['git', 'init'])
            subprocess.run(['git', 'add', '-A'])
            subprocess.run(['git', 'commit', '-m', 'Initial commit'])
            subprocess.run(['git', 'remote', 'add', 'origin', f'https://github.com/{username}/{repo_name}.git'])
            subprocess.run(['git', 'push', '-u', 'origin', 'master'])
            print("Программа загружена на GitHub")
     
        elif choice == "0":
            print("\nПрограмма завершена")
            break
        else:
            print("Некорректный выбор. Повторите попытку.")
if __name__ == "__main__":
    main()
```
